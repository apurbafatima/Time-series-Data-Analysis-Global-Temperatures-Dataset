```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/drive/MyDrive/Colab Notebooks/GLB.Ts+dSST.csv'  # Adjust as necessary for your environment
data = pd.read_csv(file_path, header=1)

# Rename columns for readability
data.columns = ["Year", "Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec",
                "J-D", "D-N", "DJF", "MAM", "JJA", "SON"]

# Convert month columns to numeric values, replacing any non-numeric entries with NaN
for col in data.columns[1:13]:  # Only the monthly columns
    data[col] = pd.to_numeric(data[col], errors='coerce')

# Calculate correlations between monthly temperature anomalies
monthly_data = data.iloc[:, 1:13]  # Select only monthly anomaly columns
correlation_matrix = monthly_data.corr()

# Plot the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", cbar_kws={'label': 'Correlation Coefficient'})
plt.title("Correlation Heatmap of Monthly Temperature Anomalies")
plt.show()
```
